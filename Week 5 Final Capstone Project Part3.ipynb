```python
import numpy as np # library to handle data in a vectorized manner
import pandas as pd # library for data analsysis
import requests
```

# Arcadia and Monterey Park School Quality Analysis

```python
# Read data from a file
df = pd.read_csv("School District.csv") 
df
```

```python
# Set district as index
df.set_index("DISTRICT", inplace=True)
df.head()
```

```python
# Transpose the above dataframe
ac_mp = df.transpose()
ac_mp
```

```python
# Check the columns types
ac_mp.dtypes
```

```python
# use the inline backend to generate the plots within the browser
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
# Horizontal bar plot of the dataframe
ac_mp.plot(kind='barh', figsize=(12, 8))

plt.xlabel('School Districts') # add to x-label to the plot
plt.ylabel('Readiness') # add y-label to the plot
plt.title('School District in Arcadia and Monterey Park Readiness') # add title to the plot

plt.show()
```

```python
# Another version of bar plot of the dataframe
ax = ac_mp.plot(kind = 'bar',
                figsize = (20,8),
                width = 0.8,
                color = ('#5cb85c', '#5bc0de', '#d9534f','#9b59b6'),
                fontsize = 14
                )

ax.set_title("School District in Arcadia and Monterey Park Readiness", fontsize = 16)

ax.legend(fontsize = 14)
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.patch.set_facecolor('white')
ax.legend(facecolor='white', framealpha=1)

ax.tick_params(left=False, labelleft=False)
for p in ax.patches:
    ax.annotate("{:.1f}".format(1*p.get_height()),
                xy=(p.get_x()+0.06, p.get_height()+1))
   # _ = ax.set_yticklabels(map('{:.0f}%'.format, ax.yaxis.get_majorticklocs()))
```
